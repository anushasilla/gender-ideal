```python
#### Adding the benchmark column

#Groups are: Financial Services, Technology, General, Non-profit
# A, B, F, J, Q, U, V financial
# C, G, M, P General
# D, H, N, O, R, S, T, X Technology
# E, I, K, L, W Non-Profit

financial_group = ["Company 1A","Company 1B","Company 1F","Company 1J","Company 1Q","Company 1U","Company 1V"]
technology_group = ["Company 1D","Company 1H","Company 1N","Company 1O","Company 1R","Company 1S","Company 1T","Company 1X"]
nonprofit_group = ["Company 1E","Company 1I","Company 1K","Company 1L","Company 1W"]

#Creating new column to contain benchmark group info
df["benchmark_group"] = np.nan

#Resetting benchmark group for finance companies 
finance = df["Organization"].isin(financial_group)
df.loc[finance,"benchmark_group"] = "Finance"

#Resetting benchmark group for technology companies 
technology = df["Organization"].isin(technology_group)
df.loc[technology,"benchmark_group"] = "Technology"

#Resetting benchmark group for technology companies 
nonprofit = df["Organization"].isin(nonprofit_group)
df.loc[nonprofit,"benchmark_group"] = "NonProfit"


#Printing out the dataframe Organization and benchmark columns to double check
#Note that any companies in general will remain as Nan as these companies visualizations
#will not be shown with a benchmark filter
df[["Organization","benchmark_group"]]
```
